```python
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import QApplication, QMainWindow

import sys
#Aqui cria-se um análogo melhor de como fazer isso usando classes
#Ver os comandos em python me lembra de como eu devo revisar linguagem
#orientada a objetos
#Aqui se cria uma janela com as devidas funcoes de evento e caracteristicas e elementos da janela
class MinhaJanela(QMainWindow):
    #Aqui estão as características da janela
    def __init__(self):
        super(MinhaJanela,self).__init__()
        self.setGeometry(500,500,300,300)
        self.setWindowTitle("MegaTeste")
        #Esses elementos são estranhamente puxados
        #antes da função de criação desses elementos ser feito
        #analisando por uma visão linear, isso deve ser por
        #causa desse self.
        self.iniciarUI()
        #Aqui estão os elementos da tela
    def iniciarUI(self):
        self.label = QtWidgets.QLabel(self)
        self.label.setText("MegaResre")
        self.label.move(50,50)

        self.botao = QtWidgets.QPushButton(self)
        self.botao.setText("EMAGTETSE")
        self.botao.clicked.connect(self.clicou)
    def clicou(self):
        self.label.setText("Você apertou o botão")
        self.atualizaLabel()
    def atualizaLabel(self):
        #Esse comando atualiza o tamanho da label
        self.label.adjustSize()
def janela():
    #Ditar as configurações da janela
    app = QApplication(sys.argv)
    jan = MinhaJanela() #Aqui chama a classe criada
    #Comando para mostrar essa janela
    jan.show()
    #Comando para parar o programa quando a janela for fechada
    sys.exit(app.exec_())
    
#Utilização dessa função
janela()
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QVBoxLayout, QLabel, QFileDialog

class App(QWidget):
    def __init__(self):
        super().__init__()
        self.title = 'PyQt5 - Selecionar Arquivo'
        self.initUI()

    def initUI(self):
        self.setWindowTitle(self.title)
        self.setGeometry(100, 100, 400, 200)
        
        layout = QVBoxLayout()

        self.label = QLabel('Caminho do arquivo selecionado aparecerá aqui', self)
        layout.addWidget(self.label)
        
        self.button = QPushButton('Selecionar Arquivo', self)
        self.button.clicked.connect(self.openFileNameDialog)
        layout.addWidget(self.button)

        self.setLayout(layout)
        self.show()

    def openFileNameDialog(self):
        options = QFileDialog.Options()
        options |= QFileDialog.ReadOnly
        fileName, _ = QFileDialog.getOpenFileName(self, "Selecionar Arquivo", "", "Todos os Arquivos (*);;Arquivos de Texto (*.txt)", options=options)
        if fileName:
            self.label.setText(fileName)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = App()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QVBoxLayout, QLabel, QFileDialog

class App(QWidget):
    def __init__(self):
        super().__init__()
        self.title = 'PyQt5 - Selecionar Pasta'
        self.initUI()

    def initUI(self):
        self.setWindowTitle(self.title)
        self.setGeometry(100, 100, 400, 200)
        
        layout = QVBoxLayout()

        self.label = QLabel('Caminho da pasta selecionada aparecerá aqui', self)
        layout.addWidget(self.label)
        
        self.button = QPushButton('Selecionar Pasta', self)
        self.button.clicked.connect(self.openDirectoryDialog)
        layout.addWidget(self.button)

        self.setLayout(layout)
        self.show()

    def openDirectoryDialog(self):
        options = QFileDialog.Options()
        options |= QFileDialog.ShowDirsOnly
        directory = QFileDialog.getExistingDirectory(self, "Selecionar Pasta", "", options=options)
        if directory:
            self.label.setText(directory)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = App()
    sys.exit(app.exec_())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
import os
import glob
import pandas as pd

def find_peaks_advanced(y):
    # Estatísticas básicas dos dados
    mean_y = np.mean(y)
    std_y = np.std(y)
    
    # Configuração adaptativa dos parâmetros
    height = mean_y + std_y  # Altura mínima como uma função da média e desvio padrão
    prominence = std_y  # Proeminência mínima como uma função do desvio padrão
    distance = int(len(y) / 100)  # Distância mínima baseada no tamanho dos dados
    width = int(len(y) / 200)  # Largura mínima baseada no tamanho dos dados
    
    # Detecta picos usando parâmetros adaptativos
    peaks, properties = find_peaks(y, height=height, prominence=prominence, distance=distance, width=width)
    
    return peaks, properties

# Dados de exemplo
caminho=r"C:\Users\aojor\Downloads\CodigosPython\vsCode\projetos\pacoteComparadorPicos\comparadorPicos"
buscaXyPadrao=os.path.join(caminho,'*.xy')
ArrayCaminhoXy=glob.glob(buscaXyPadrao)
dataFramePadrao =pd.read_csv(ArrayCaminhoXy[0], delim_whitespace=True, header=None, names=['x','y'])
angulos=dataFramePadrao.x.values
intensidades=dataFramePadrao.y.values

# Encontra picos
peaks, properties = find_peaks_advanced(intensidades)

# Plot dos dados e picos detectados
plt.plot(angulos, intensidades, label='Dados de Difração')
plt.plot(angulos[peaks], intensidades[peaks], 'rx', label='Picos Detectados')
plt.xlabel('Ângulo 2θ')
plt.ylabel('Intensidade')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import os
import glob
import pandas as pd
caminho=r"C:\Users\aojor\Downloads\CodigosPython\vsCode\projetos\pacoteComparadorPicos\comparadorPicos"
buscaXyPadrao=os.path.join(caminho,'*.xy')
ArrayCaminhoXy=glob.glob(buscaXyPadrao)
# Carregar dados do arquivo .xy
data = np.loadtxt(ArrayCaminhoXy[0])
x = data[:, 0]
y = data[:, 1]

# Detectar picos
peaks, properties = find_peaks(y, height=600, distance=10, prominence=0.05)

# Plot dos dados e picos detectados
plt.plot(x, y, label='Dados de Difração')
plt.plot(x[peaks], y[peaks], 'rx', label='Picos Detectados')
plt.xlabel('Ângulo 2θ')
plt.ylabel('Intensidade')
plt.legend()
plt.show()
```
